**Problem Statement 1:** In a survey conducted by a non-banking financial company, a sample of 200 customers yielded that x of them were highly satisfied with the timely disbursal of their loans.

Write a Python code to perform the following operations:
1. Read an integer input that specifies the number of highly satisfied customers
2. Calculate an approximate 90% confidence interval for the proportion of the loan customers who are highly satisfied with disbursal time
  * Find out the Margin of Error using scipy.stats.norm.ppf
  * Calculate and print the confidence interval values rounded up to five decimal places and separated by a space

>**Note:**
* Margin of Error = Critical Value
* Standard Error of Statistic
* Confidence Interval = Sample Statistic ± Margin of Error

\
**Example:** Let's say 172 out of 200 customers were highly satisfied.

**Sample Input:**
172

The confidence interval should be printed as -

**Sample Output:**
0.82856 0.89144

```python
# imports
import numpy as np
import scipy.stats as stats
import math as m
```

```python
# Read an integer input that specifies the number of highly satisfied customers
num_highly_satisfied = int(input("Enter the number of highly satisfied customers: "))

# According to data
total_customers = 200

# Calculate the proportion of highly satisfied customers
sample_proportion = num_highly_satisfied / total_customers

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / total_customers)

# Calculate the 90% confidence level for the proportion of satisfied customers
margin_of_error = stats.norm.ppf(0.90) * (standard_error)

# # Calculate the lower and upper bounds of the confidence interval
CI_lower_bound, CI_upper_bound = sample_proportion - margin_of_error, sample_proportion + margin_of_error

# # Print the confidence interval and margin of error
print(f"Confidence Interval: {np.round(CI_lower_bound, 5)} {np.round(CI_upper_bound, 5)}")
```

**Problem Statement 2:** A radar unit is used to measure the speeds of cars on a motorway. The speeds are normally distributed with a mean of 75 km/hr and a standard deviation of 15 km/hr.

Write a Python code to perform the following operations:

1. Find the probability that a car picked at random is traveling at more than X km/hr
  
  * Take the speed X as an input
  * Print the probability value rounded up to four decimal places

Hint: Use Normal Distribution.

**Sample Input:**
100

**Sample Output:**
0.0478

```python
# Mean of the speeds and standard deviation given in data
mean_speed = 75
standard_deviation = 15

# Read the input speed x
X = float(input("Enter the input speed: "))

# Calculate the probability that a car picked at random is traveling at more than X km/hr
probability_more_than_X = 1 - stats.norm.cdf(X, loc=mean_speed, scale=standard_deviation) # CDF gives the probability that the random variable X is less than or equal to a particular value x.

# Print the probability value rounded up to four decimal places
print("Probability that a car picked at random is traveling at more than", X, "km/hr:", round(probability_more_than_X, 4))
```

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a Datarame with columns –“YEAR”, “JUN_GT_500” (Contains a boolean value to show whether it rained more than 500 mm in the month of June
df = pd.DataFrame(data['YEAR'])
df['JUN_GT_500'] = (data['JUN'] > 500).astype('int')

# “JUL_GT_500” (Contains a boolean value to show whether it rained more than 500 mm in the month of July),
df['JUL_GT_500'] = (data['JUL'] > 500).astype('int')

# “FLOODS” (Contains a boolean value to show whether it flooded that year)
df['FLOODS'] = (data['FLOODS'] == 'YES').astype('int')

# Create a count column in the dataframe
df['COUNT'] = 1
df.head()
```

##### Calculate the individual probabilities for JUNE

```python
# 1. Calculate the probability of rainfall greater than 500 in
june = df.groupby('JUN_GT_500').count()['COUNT'][1]

# 2. Probabiltiy of floods happening in the year
floods = df.groupby('FLOODS').count()['COUNT'][1]

# 3. Probabilty of rainfall greater than 500mm in June given that it flood, P(JUN_GT_500|FLOOD)
june_flood = df.groupby(['JUN_GT_500','FLOODS']).count()['COUNT'][1,1]
juneflood = june_flood/floods
```

```python
# Calculate the probability of flood given it rained more than 500 mm in June (P(FLOODS|JUN_GT_500))
flood_given_june = (floods * juneflood)/june
print(f"The probability of flood given it rained more than 500 mm in June:\n\n (P(Flood|June))= {np.round(flood_given_june, 5)}")
```

```python
# Calculate the probability of rain more than 500 mm in June, given it flooded that year (P(JUN_GT_500|FLOODS))
print(f"The probability of flood given it rained more than 500 mm in June:\n\n (P(June|Flood))= {june_flood/floods}")
```

##### Calculate the individual probabilities for JULY

```python
# 1. Calculate the probability of rainfall greater than 500 in JULY
july = df.groupby('JUL_GT_500').count()['COUNT'][1]

# 2. Probabilty of rainfall greater than 500mm in July given that it flood, P(JUL_GT_500|FLOOD)
july_flood = df.groupby(['JUL_GT_500','FLOODS']).count()['COUNT'][1,1]
julyflood = july_flood/floods
```

```python
# Probability of flood given it rained more than 500 mm in July (P(FLOODS|JUL_GT_500))
flood_given_july = (floods * julyflood)/july
print(f"The probability of flood given it rained more than 500 mm in July:\n\n (P(Flood|July))= {np.round(flood_given_july, 5)}")
```

```python
# Probability of rain more than 500 mm in July given it flooded that year (P(July|Flood))
print(f"The probability of flood given it rained more than 500 mm in June:\n\n (P(July|Flood))= {july_flood/floods}")
```

**Problem Statement 4**: Write a Python program to load the wine dataset using the Sklearn library to a DataFrame and perform the following tasks:
1. Convert the dataset into DataFrame using pandas.
2. Generate the sample size of 50 and give a random state as 100.
3. Calculate Z-critical, Margin of Error, and Confidence Interval for alcohol at 95% significance interval on generated sample data.

```python
# Imports
from sklearn.datasets import load_wine
import scipy.stats as stats
```

```python
# Load the dataset from the sklearns datasets module
wine_data = load_wine()

# Convert the dataset into dataframe
wine_df = pd.DataFrame(wine_data['data'], columns = wine_data['feature_names'])
wine_df['target'] = wine_data.target

wine_df.head()
```

```python
# Generate the sample size of 50 and give a random state as 100.
sample_data = wine_df.sample(n = 50, random_state = 100)
len(sample_data)
```

```python
# Calculate the standard_deviation and mean for sample data
alcohol_mean = sample_data['alcohol'].mean()
alcohol_std_dev = sample_data['alcohol'].std()

alcohol_mean, alcohol_std_dev
```

```python
# Calculate Z-critical, Margin of Error, and Confidence Interval for alcohol on generated sample data
# set 95% significance interval
alpha = 0.95

# Calculate Z-critical
z_critical = stats.norm.ppf(alpha)

# Calculate Margin of Error
error_margin = z_critical * (alcohol_std_dev/np.sqrt(len(sample_data)))

# Calculate Confidence intervals
confidence_interval = (alcohol_mean - error_margin, alcohol_mean + error_margin)

# Print the values
print("Z-critical value:", z_critical)
print("Margin of Error:", error_margin)
print("Confidence Interval:", confidence_interval)
```
