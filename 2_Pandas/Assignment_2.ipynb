Problem Statement 1: Samantha has created a dataset named 'top50spotify.csv' of her top 50 songs from spotify.

Dataset Description:top50spotify.csv - The dataset contains 14 features. Here's a brief description of a few columns in the dataset:
• SerialNo. - Serial number of songs
• Track.Name - Name of the track
• Artist.Name - Name of the artist
• Genre - Genre of the song
• Energy - Energy index of the song
• Length. - Length of the song
• Popularity - Popularity index of the song

Tasks to be performed:

1. Import the dataset as a DataFrame and drop the first column.
2. Save it as 'top50.csv'.
3. Find the average Energy and Length of first 10 songs.
4. Find the total length of songs, group by genre from top to bottom.
5. Print the artist name with the most number of tracks in one genre. (Hint: Group by artist name and genre)
6. Print the data of the tracks created by the artist from the previous question.

Note: Fetch and download the dataset using the following link https://www.dropbox.com/s/2hg67jin2n852mz/top50spotify.csv

```python
!wget https://www.dropbox.com/s/2hg67jin2n852mz/top50spotify.csv
```

```python
# Import the necessary libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data using pandas and drop the first column
top50 = pd.read_csv('top50spotify.csv').iloc[:,1:]
top50.head()
```

```python
# Save the data as top50 csv
top50.to_csv('top50.csv')
```

```python
# Find the average energy and length of first 10 songs
means = top50[['Energy', 'Length.']].head(10).mean()

# Plot the bar plot for the mean values
ax = means.plot(kind = 'bar', color = ['salmon','lightblue'])

# Annotate each bar with its corresponding mean value
for i, v in enumerate(means):
    ax.text(i, v + 0.05, str(round(v, 2)), ha='center', va='bottom')

plt.show()
```

```python
# Find the total length of the songs, group by genre from top to bottom
top50.groupby('Genre')['Length.'].sum().sort_values(ascending = True)
```

```python
# Print the artist name with the most number of tracks in one genre
artist = top50.groupby(['Artist.Name', 'Genre'])['Track.Name'].count().idxmax()
print(f'The artist with most tracks in one genre is :\nArtist : {artist[0]} \nGenre : {artist[1]}')
```

```python
# Print the data of the tracks created by the artist from the previous question.
top50[top50['Artist.Name'] == artist[0]]
```

Problem Statement 2: Write a Python program to perform the following tasks-

1. Create a pandas series from the below dictionary where indices are subjects:

{'English':{'Sam':60,'Jackson':74,'Ahree':85},

'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},

'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},

'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},

'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}
}

2. Convert the created series into DataFrame and replace the null values with zeroes.

3. Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.

```python
# Create a pandas series from the below dictionary where indices are subjects
series = pd.Series({'English':{'Sam':60,'Jackson':74,'Ahree':85},
                    'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
                    'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
                    'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
                    'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}})

# Check the indices of the series
series.index
```

```python
# Convert the created series into DataFrame and replace the null values with zeroes.
df = pd.DataFrame.from_records(series, index = series.index)

# Replace null values with zeroes
df.fillna(0, inplace=True)

df
```

```python
# Calculate the average of each subject
average = [np.round(df[column].mean(), 2) for column in df.columns]

# Transpose the DataFrame, create a new column 'Average' and fill the values in it by calculating the average of all subjects.
df_new = df.transpose()
df_new['average'] = average

df_new
```

Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17.

```python
# Program to create a series from 1 to 1000 with numbers divisible by 7 and 17.
num_series = pd.Series([num for num in range(1000) if (num % 7 == 0 and num % 17 == 0)])
num_series
```

Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.

Dataset Description: cereal.csv - The dataset contains 16 features.
Here's a brief description of 3 columns in the dataset:
* name - Brand name of the cereals
* MFR - Manufacturer of the brands
* rating - Rating of the cereals

Syliphia wants to visualize the quality of cereals and determine which manufacturer delivers the best quality.

* Tasks to be performed:
 1. Import the dataset.
 2. Plot ratings of different types of manufacturers.
 3. Use xticks range form 0-100.
 4. Change the style of the graph to seaborn Note:

Fetch and download the dataset using the following link https://www.dropbox.com/s/idnul34dfo5cnke/cereal.csv

```python
!wget https://www.dropbox.com/s/idnul34dfo5cnke/cereal.csv
```

```python
# Inport the dataset
cereal = pd.read_csv('cereal.csv')
cereal.head()
```

```python
# Plot the ratings of different types of manufacturers
mfg_ratings = pd.DataFrame(cereal.groupby('mfr')['rating'].mean())

sns.set_style('darkgrid')
plt.figure(figsize = (8,5))
plt.barh(mfg_ratings.index, mfg_ratings.rating, color = 'steelblue', label = 'ratings')
plt.xlabel('Rating')
plt.ylabel('Mfr')
plt.xticks(range(0, 100, 10))
plt.title('Ratings of different types of manufacturers')
plt.legend()
plt.show()
```
